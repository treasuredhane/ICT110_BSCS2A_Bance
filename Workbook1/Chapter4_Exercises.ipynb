Rewrite your pay computation with time-and-a-half for overtime and create a function called compute pay which takes two parameters (hours and rate).

```python
Enter Hours: 45
Enter Rate: 10
Pay: 475.0
```

[PEW] E4.6

```python
def compute_pay(hours, rate):
    if hours <= 40:
        pay = hours * rate
    else:
        regular_pay = 40 * rate
        overtime_hours = hours - 40
        overtime_pay = overtime_hours * (rate * 1.5)
        pay = regular_pay + overtime_pay
    return pay

def main():
    hours = float(input("Enter the number of hours worked: "))
    rate = float(input("Enter the hourly rate: "))
    total_pay = compute_pay(hours, rate)
    print("Total pay is:", total_pay)

if __name__ == "__main__":
    main()
```

Challenge

Write a function called computegrade that takes a score as its parameter and returns a grade as a string using the following table:

```python
Score           Grade
[ 97, 100]      1.0
[ 93,  97)        1.25
[ 89,  93)        1.5
[ 85,  89)        1.75
[ 81,  85)        2.0
[ 77,  81)        2.25
[ 74,  77)        2.5
[ 0,   74)        3.0

Run the program repeatedly to test the various different values for input.
```

Adapted from [PEW] E4.6

```python
def computegrade(score):
    if 97 <= score <= 100:
        return '1.0'
    elif 93 <= score < 97:
        return '1.25'
    elif 89 <= score < 93:
        return '1.5'
    elif 85 <= score < 89:
        return '1.75'
    elif 81 <= score < 85:
        return '2.0'
    elif 77 <= score < 81:
        return '2.25'
    elif 74 <= score < 77:
        return '2.5'
    elif 0 <= score < 74:
        return '3.0'
    else:
        return 'Invalid score'

# Test the function with different values for input
scores = [100, 95, 90, 87, 83, 79, 75, 72, -5, 105]
for score in scores:
    print(f"Score: {score}, Grade: {computegrade(score)}")
```
