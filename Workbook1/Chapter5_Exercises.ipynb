Investigate on the difference between is and == given the following variables.

```python
a = 2
b = 2.0
c = a
```

```python
a = 2
b = 2.0
c = a

print(a == b)  # Output: True (values are equal)
print(a is b)  # Output: False (different types, different memory locations)

print(a == c)  # Output: True (values are equal)
print(a is c)  # Output: True (same value, same memory location)
```

Challenge 1

Write a program which repeatedly reads numbers until the user enters "done". Once "done" is entered, print out the total, count, and average of the numbers. If the user enters anything other than a number, detect their mistake using try and except and print an error message and skip to the next number.

```python
Enter a number: 4
Enter a number: 5
Enter a number: bad data
Invalid input
Enter a number: 7
Enter a number: done
16 3 5.333333333333333
```

```python
total = 0
count = 0

while True:
    user_input = input("Enter a number (or 'done' to finish): ")

    if user_input.lower() == 'done':
        break

    try:
        number = float(user_input)
        total += number
        count += 1
    except ValueError:
        print("Invalid input")
        continue

if count > 0:
    average = total / count
    print(total, count, average)
else:
    print("No numbers entered.")
```

Challenge 2

Write another program that prompts for a list of numbers as above and at the end prints out both the maximum and minimum of the numbers instead of the average.

[PEW] E5.2

```python
def main():
    numbers = []  # Store entered numbers

    while True:
        user_input = input("Enter a number or 'done' to finish: ")

        if user_input.lower() == 'done':
            break

        try:
            number = float(user_input)
            numbers.append(number)
        except ValueError:
            print("Invalid input. Please enter a number or 'done'.")
            continue

    if numbers:
        print("Maximum:", max(numbers))
        print("Minimum:", min(numbers))
    else:
        print("No numbers entered.")

if __name__ == "__main__":
    main()
```
