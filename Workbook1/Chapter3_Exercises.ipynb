Exercise 2

Write your salary computation to give an employee 1.5 times the hourly rate for hours worked above 40 hours. Example:

```python
Enter Hours: 45
Enter Rate: 10
Pay: 475.0
```

[PEW] E3.2

```python
def calculate_salary(hourly_rate, hours_worked):
    regular_hours = min(hours_worked, 40)
    overtime_hours = max(hours_worked - 40, 0)

    regular_pay = regular_hours * hourly_rate
    overtime_pay = overtime_hours * (hourly_rate * 1.5)

    total_pay = regular_pay + overtime_pay
    return total_pay

hourly_rate = float(input("Enter the hourly rate: "))
hours_worked = float(input("Enter the number of hours worked: "))

salary = calculate_salary(hourly_rate, hours_worked)
print("The total salary is:", salary)
```

CHALLENGE

Write a program to prompt for a score between 0 and 100 and print a grade using the following table:

```python
Score           Grade
[ 97, 100]      1.0
[ 93,  97)        1.25
[ 89,  93)        1.5
[ 85,  89)        1.75
[ 81,  85)        2.0
[ 77,  81)        2.25
[ 74,  77)        2.5
[ 0,   74)        3.0
```

```python
def calculate_grade(score):
    if score >= 97:
        return 1.0
    elif score >= 93:
        return 1.25
    elif score >= 89:
        return 1.5
    elif score >= 85:
        return 1.75
    elif score >= 81:
        return 2.0
    elif score >= 77:
        return 2.25
    elif score >= 74:
        return 2.5
    else:
        return 3.0

def main():
    score = float(input("Enter the score (between 0 and 100): "))
    if 0 <= score <= 100:
        grade = calculate_grade(score)
        print("The grade for the score {} is: {}".format(score, grade))
    else:
        print("Invalid score. Please enter a score between 0 and 100.")

if __name__ == "__main__":
    main()
```
